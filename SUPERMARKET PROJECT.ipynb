```python
# Convert date if not already
df['Date'] = pd.to_datetime(df['Date'])

# Daily income trend
daily_income = df.groupby('Date')['gross income'].sum()

daily_income.plot(title="Daily Gross Income Trend")
plt.ylabel("Gross Income")
plt.xlabel("Date")
plt.show()
```

```python
income_by_product = df.groupby('Product line')['gross income'].sum().sort_values(ascending=False)
print(income_by_product)

income_by_product.plot(kind='bar', title='Gross Income by Product Line')
plt.ylabel("Gross Income")
plt.show()
```

```python
spending_gender = df.groupby(['Gender', 'Product line'])['Total'].sum().unstack()

print(spending_gender)

# Visualization
spending_gender.plot(kind='bar', stacked=False, figsize=(12,6))
plt.title("Spending Pattern by Gender and Product Line")
plt.ylabel("Total Spending")
plt.xticks(rotation=45)
plt.show()
```

```python
# Ensure Date is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract weekday name
df['Weekday'] = df['Date'].dt.day_name()

# Sum sales by weekday
sales_by_weekday = df.groupby('Weekday')['Total'].sum()

# To get weekdays in order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_weekday = sales_by_weekday.reindex(weekday_order)

print(sales_by_weekday)

# Plot
sales_by_weekday.plot(kind='bar', figsize=(10,5))
plt.title("Total Sales by Day of Week")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_by_hour = df.groupby('Time')['Total'].sum()

sales_by_hour.plot(kind='bar', figsize=(12,6))
plt.title("Total Sales by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Total Sales")
plt.show()
```

```python
income_by_product_line = df.groupby('Product line')['gross income'].sum().sort_values(ascending=False)

print(income_by_product_line)

income_by_product_line.plot(kind='bar', figsize=(10,6))
plt.title("Gross Income by Product Line")
plt.ylabel("Gross Income")
plt.show()
```

```python
# Total gross income by city
income_by_city = df.groupby('City')['gross income'].sum().sort_values(ascending=False)
print("Gross Income by City:")
print(income_by_city)

# Focus city
best_city = income_by_city.idxmax()
print(f"\nCity for expansion: {best_city}")

# Product performance in best city
income_city_product = df[df['City'] == best_city].groupby('Product line')['gross income'].sum().sort_values(ascending=False)
print(f"\nProduct performance in {best_city}:")
print(income_city_product)

# Visualization
income_city_product.plot(kind='bar', figsize=(10,6))
plt.title(f"Gross Income by Product Line in {best_city}")
plt.ylabel("Gross Income")
plt.show()
```
